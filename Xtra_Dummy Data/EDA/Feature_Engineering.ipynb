```python
# import the libraries needed
import pandas as pd
import numpy as np
```

```python
# Load the dataset
path = "../sample_COPD_dataset.csv"
df = pd.read_csv(path)
df.head()
```

```python
# Binning Continuous Variables
bins = [39, 49, 59, 69, 79]
labels = ['40-49', '50-59', '60-69', '70-79']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
# Interaction between Age and Tobacco Smoking
df['Age_Smoking_Interaction'] = df['Age'] * df['Tobacco Smoking']

# Interaction between Income Level and Biomass Fuel
df['Income_Biomass_Interaction'] = df['Income Level'].map({'Low': 1, 'Middle': 2, 'High': 3}) * df['Biomass Fuel']
```

```python
# One-hot encoding for categorical variables
df = pd.get_dummies(df, columns=['Location', 'Education Level', 'Occupation', 'Age Group'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

# List of numeric columns to scale
numeric_columns = ['Age', 'Outdoor Air Poll', 'Occupational Expos', 'Respiratory Infection']

scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
```

```python
# Save the new DataFrame
df.to_csv("COPD_feature_engineered.csv", index=False, encoding='utf-8')
```
